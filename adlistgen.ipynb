```python
import xml.etree.ElementTree as et
import xml.dom.minidom as md
import sys
# parameters
adhesion_index=1
rootname='bind_tableBRA' + str(adhesion_index) + 'KET'
total_cell=128
vartical_number=16
horizontal_number=int(total_cell/vartical_number)
# root generation
xml_root = et.Element(rootname)
# cell loop
for cell_index in range(total_cell):
    parent_name='cellBRA' + str(cell_index) + 'KET'
    parent = et.SubElement(xml_root, parent_name)
    line_number=int(cell_index/vartical_number)
#    line = et.SubElement(parent, "line")
#    line.text = str(line_number)
    if line_number%2 == 0 :
        k=cell_index-vartical_number-1
        if k >= 0 and int(k/vartical_number) == line_number-1:
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        elif cell_index%vartical_number == 0:
            k = cell_index-1
            if k >= 0:
                child = et.SubElement(parent, 'component')
                child.text = str(k)
        k=cell_index-vartical_number
        if k >= 0 and int(k/vartical_number) == line_number-1:
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        k=cell_index+1
        if int(k/vartical_number) == line_number and k < total_cell:
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        elif cell_index%vartical_number == vartical_number-1:
            k = cell_index - vartical_number+1
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        k=cell_index-1
        if int(k/vartical_number) == line_number and k >= 0 :
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        elif cell_index%vartical_number == 0:
            k = cell_index + vartical_number-1
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        k=cell_index+vartical_number-1
        if k < total_cell and int(k/vartical_number) == line_number+1:
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        elif cell_index%vartical_number == 0:
            k = cell_index + 2*vartical_number-1
            if k < total_cell :
                child = et.SubElement(parent, 'component')
                child.text = str(k)
        k=cell_index+vartical_number
        if k < total_cell and int(k/vartical_number) == line_number+1:
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        elif cell_index%vartical_number == vartical_number-1:
            k = cell_index + 1
            if k < total_cell :
                child = et.SubElement(parent, 'component')
                child.text = str(k)
    else:
        k=cell_index-vartical_number
        if k >= 0 and int(k/vartical_number) == line_number-1:
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        k=cell_index-vartical_number+1
        if k >= 0 and int(k/vartical_number) == line_number-1:
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        elif cell_index%vartical_number == vartical_number-1:
            k = cell_index-2*vartical_number+1
            if k >= 0:
                child = et.SubElement(parent, 'component')
                child.text = str(k)
        k=cell_index+1
        if int(k/vartical_number) == line_number and k < total_cell:
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        elif cell_index%vartical_number == vartical_number-1:
            k = cell_index - vartical_number+1
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        k=cell_index-1
        if int(k/vartical_number) == line_number and k < total_cell:
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        elif cell_index%vartical_number == vartical_number-1:
            k = cell_index - vartical_number+1
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        k=cell_index+vartical_number
        if k < total_cell and int(k/vartical_number) == line_number+1:
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        k=cell_index+vartical_number+1
        if k < total_cell and int(k/vartical_number) == line_number+1:
            child = et.SubElement(parent, 'component')
            child.text = str(k)
        elif cell_index%vartical_number == vartical_number-1:
            k = cell_index + 1
            if k < total_cell :
                child = et.SubElement(parent, 'component')
                child.text = str(k)
#
# output part
string = et.tostring(xml_root, 'utf-8')
document = md.parseString(string).toprettyxml(indent='   ')
with open('test.txt', 'w') as file:
    file.write((document.replace('BRA', '[').replace('KET', ']')))
```
